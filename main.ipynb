```julia
function examine_seed_pool(N :: Int, M :: Int, range :: Base.RangeIndex{Int})
    env = Gurobi.Env()
    results = []
    for seed in range
        data = random_data(N, M, seed = seed)
        obj, x, y = solve_FL(data, relaxed = false, env = env)
        obj_rel, x_rel, y_rel, _, _ = solve_FL(data, relaxed = true, env = env)
        push!(results, (seed, obj - obj_rel, x == x_rel, y == y_rel))
    end

    return results
end
```

```julia
res = examine_seed_pool(50, 25, 1:1000);
```

```julia
# From the list of seeds, generate a md table with columns :
# Seed | N | M | Relaxed obj | Integer obj | Rounding Value | Primal-Dual Value

function compare_heuristics(N :: Int, M :: Int, seeds :: Vector{Int}) :: Vector{Tuple{Int, Int, Int, Float64, Int, Int, Int}}
    env = Gurobi.Env()
    results = []
    for seed in seeds
        data = random_data(N, M, seed = seed)
        obj_rel, x_rel, y_rel = solve_FL(data, relaxed = true, env = env)
        obj_int, x_int, y_int = solve_FL(data, relaxed = false, env = env)
        cost_rounding, _, _ = rounding_heuristic(data, env = env)
        cost_primal_dual, _, _ = primal_dual_heuristic(data)

        push!(results, (seed, N, M, obj_rel, obj_int, cost_rounding, cost_primal_dual))
    end

    return results
end
```

```julia
N1 = 25
M1 = 10
SEEDS1 = [77, 735, 776, 908, 1, 2, 3, 4, 5, 6]
res1 = compare_heuristics(N1, M1, SEEDS1);
```

```julia
N2 = 50
M2 = 25
SEEDS2 = [41, 88, 143, 159, 204, 226, 294, 330, 333, 377, 468, 499, 504, 599, 624, 653, 685, 5, 6, 7]
res2 = compare_heuristics(N2,M2,SEEDS2);
length(SEEDS2)
```

```julia
res = vcat(res1, res2);
```

```julia

# Output the results in a markdown table in a file 
function to_md_table(
    results :: Vector{Tuple{Int, Int, Int, Float64, Int, Int, Int}}, 
    filename :: String
    )
    open(filename, "w") do io
        println(io, "| Instance | N | M | Relaxed obj | Integer obj | Integrality Gap | Rounding heuristic | Primal-Dual Heuristic |")
        println(io, "| :----: | -: | -: | -: | -: | -: | -: | -: |")
        for (i, (seed, N, M, obj_rel, obj_int, cost_rounding, cost_primal_dual)) in enumerate(results)
            integrality_gap = obj_int - obj_rel
            println(io, "| $i | $N | $M | $obj_rel | $obj_int | $integrality_gap | $cost_rounding | $cost_primal_dual |")
        end
    end
end
```

```julia
to_md_table(res, "results.md")
```

```julia
include("src/draw.jl")

# Draw a scatter plot on the instance with the "best" seed
example = random_data(50, 25, seed = 143)
obj_exact, y_exact, x_exact = solve_FL(example, relaxed = false)

@assert all(y_exact .== 0 .|| y_exact .== 1)
@assert all(x_exact .== 0 .|| x_exact .== 1)
@assert all(sum(x_exact, dims = 2) .== 1)

opened_facilities_exact = findall(y_exact .== 1)
client_assignment_exact = [findfirst(x_exact[i, :] .== 1) for i in 1:example.N];

@assert all(client_assignment_exact[i] in opened_facilities_exact for i in 1:example.N)
@assert all(length(findall(x_exact[:, j] .== 1)) > 0 for j in opened_facilities_exact)
```

```julia
draw_solution(example, opened_facilities_exact, client_assignment_exact, figname = "Exact", cost = obj_exact)
```

```julia
obj_rounding, opened_facilities_rounding, client_assignment_rounding = rounding_heuristic(example)

@assert all(client_assignment_rounding[i] in opened_facilities_rounding for i in 1:example.N)
@assert all(length(findall(client_assignment_rounding .== j)) > 0 for j in opened_facilities_rounding)

draw_solution(example, opened_facilities_rounding, client_assignment_rounding, figname = "Rounding", cost = obj_rounding)
```
